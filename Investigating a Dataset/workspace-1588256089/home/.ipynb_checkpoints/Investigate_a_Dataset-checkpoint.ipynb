### Below are the columns description of the dataset:
 - id - A unique identifier for each movie.
 - imdb_id - Imdb_id is also an identifier for each movie.
 - cast - The name of lead and supporting actors.
 - director - The name of Director.
 - budget - The budget in which the movie was made.
 - genre - The genre of the movie, Action, Comedy ,Thriller etc.
 - homepage - A link to the homepage of the movie.
 - keywords - The keywords or tags related to the movie.
 - original_title - The title of the movie before translation or adaptation.
 - overview - A brief description of the movie.
 - popularity - A numeric quantity specifying the movie popularity.
 - production_companies - The production house of the movie.
 - release_date - The date on which it was released.
 - revenue - The worldwide revenue generated by the movie.
 - runtime - The running time of the movie in minutes.
 - status - "Released" or "Rumored".
 - tagline - Movie's tagline.
 - vote_average - average ratings the movie recieved.
 - vote_count - the count of votes recieved.

### Q1. What are the correaltions between the feature variables in this dataset?

```python
plt.figure(figsize=(16,8))
sns.heatmap(df.corr(),square=True,annot=True);
```

#### Heatmap shows the pearson's correlation of the numeric data types.
We can clearly notice that there are features with high correlation like
 - budject_adj has a correlation of 0.97 with budget column
 - revenue_adj has a correlation of 0.92 with revenue column
 - also budget and revenue have a high correlation of 0.73, so does vote_count and revenue with 0.79
Since budget_adj and revenue_adj are highly correlated with budget and revenue respectively, we take only one of these in account for further analysis.
 

```python
df.plot(kind = 'scatter',x='vote_average',y='budget', figsize=(15,8));
```

#### Vote_average lies in the range from 0 to 10. We can notice that most of the high budget movies(more than 20000000 USD) have a vote of 6 and above. The graph doesnt clearly show any linear relation between the vote and the budget.

```python
df.plot(kind = 'scatter',x='vote_average',y='popularity',figsize=(15,8));
```

#### The vote and popularity also shows a weak relationship and it can be notices that it is not necessary for a highly voted movie to be very popular. Spectators tend to rate the movie higher evethough it is not as popular as other movies. Spectator count may be a consideration here.

```python
df.plot(kind = 'scatter',x='vote_average',y='revenue',figsize=(15,8));
```

#### It is clear from the above visual that a movie with high vote tends to generate more revenue as compared to movies with low vote. The graph is more concentrated towards high voting average.

### Q2. Is their any trend between the number of spectators voting for a partcular movie and revenue generated?

```python
df.plot(kind = 'scatter',x='vote_count',y='revenue',figsize=(15,8));
```

#### The number of spectators(vote_counts) viewing a movie has a linear relationship with the revenue generated by the movie. It a a great point which means that a more a movie is rated, the better chance of generating higher revenues.

### Q3. Is their any trend between the number of spectators voting for a particular movie and the popularity of the movie?

```python
df.plot(kind = 'scatter',x='vote_count',y='popularity',figsize=(15,8));
```

#### The number of spectators also has a linear relationship with the popularity of the movie, but it can also be noted that the slope of the linear line is less steeper as compared with revenue.

```python
df.plot(kind = 'scatter',x='popularity',y='revenue',figsize=(15,8));
```

#### It can be seen that the popularity of a movie and revenue it generates has a high linear relationship. The more popular a movie is, higher the revenue it generates.

```python
df['runtime'].hist(bins=100,figsize=(15,8));
plt.xlim(0,250)
plt.xlabel('Runtime')
plt.ylabel('frequency')
```

#### It is seen that a higher number movies that are released have a runtime in range of 80 to 120 minutes. Fewer movies are released with a runtime more or less than that range. It is normally distributed curve.

```python
df['release_year'].plot(kind='hist', bins= 30,figsize=(15,8))
plt.xticks(np.arange(1960,2016,2),rotation=90)
plt.show()
```

#### It is clear from the histogram that as years passed, there were more movies being released. It is left skewed curve with depicts that a higher number of movies were released during recent time.

```python
df['vote_average'].plot(kind='hist', bins= 30,figsize=(15,8))
plt.show()
```

#### It can be seen from the curve that more number of the movies were voted above 5 and below 7.
